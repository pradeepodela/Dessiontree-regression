```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:/Users/PRADEEP/Desktop/ml  udemy/Data Files\Data Files/3. ST Academy - Decision Trees resource files/Movie_regression.csv', header=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
 df['Time_taken'].fillna(value = df['Time_taken'].mean(),inplace=True)
```

```python
df = pd.get_dummies(df,columns=['Genre','3D_available'],drop_first=True)
```

```python
df.head()
```

#  X-Y

```python
x = df.loc[:,df.columns!='Collection']
```

```python
x.head()
```

```python
y = df['Collection']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
 x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import tree
```

```python
regtree = tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(x_train,y_train)
```

```python
x_pred_reg = regtree.predict(x_test)
```

```python
x_pred_reg.round()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn. metrics import r2_score
```

```python
r2_score(y_test,x_pred_reg)
```

```python
mean_squared_error(y_test,x_pred_reg)
```

# tree purning using max depth 

```python
regtree = tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(x_train,y_train)
x_pred_reg = regtree.predict(x_test)
plt.figure(figsize=(15,10))
tree.plot_tree(regtree,filled=True)
r2_score(y_test,x_pred_reg)
```

# tree purning using samples split


```python
regtree1 = tree.DecisionTreeRegressor(min_samples_split=40)
regtree1.fit(x_train,y_train)
x_pred_reg = regtree1.predict(x_test)
plt.figure(figsize=(20,20))
tree.plot_tree(regtree1,filled=True)
r2_score(y_test,x_pred_reg)
```

# tree purning using sample leaf


```python
regtree1 = tree.DecisionTreeRegressor(min_samples_leaf=25)
regtree1.fit(x_train,y_train)
x_pred_reg = regtree1.predict(x_test)
plt.figure(figsize=(20,20))
tree.plot_tree(regtree1,filled=True)
r2_score(y_test,x_pred_reg)
```
